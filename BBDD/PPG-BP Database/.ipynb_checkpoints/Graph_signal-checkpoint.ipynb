```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ppg_part1 = np.loadtxt('./Data File/0_subject/6_1.txt')
ppg_part2 = np.loadtxt('./Data File/0_subject/6_2.txt')
ppg_part3 = np.loadtxt('./Data File/0_subject/6_3.txt')
ppg_data = np.concatenate((ppg_part1, ppg_part2, ppg_part3))

# Graficar la señal PPG
plt.figure(figsize=(10, 4))
plt.plot(ppg_data)
plt.title('Señal PPG')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.show()
```
